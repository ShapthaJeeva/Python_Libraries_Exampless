```python
Table1 = pd.DataFrame({'CustID':['1001','1002','1003','1004'],'Q1':['101','102','103','104'],'Q2':['201','202','203','204']})

Table2 = pd.DataFrame({'CustID':['1001','1006','1003','1004'],'Q3':['301','302','303','304'],'Q4':['401','402','403','404']})
```

```python
Table1
```

```python
Table2
```

```python
# Merging
pd.merge(Table1,Table2,how='outer',on='CustID')  # all data will be merged
```

```python
pd.merge(Table1,Table2,how='inner',on='CustID')  # only those have common data will be merged
```

```python
# merge => uses common columns to combine two dataframes
# join => uses the row index to join two dataframes

Table1 = pd.DataFrame({'Q1':['101','102','103'],'Q2':['201','202','203']},index=['I0','I1','I2'])

Table2 = pd.DataFrame({'Q3':['301','302','303'],'Q4':['401','402','403']},index=['I0','I2','I3'])
```

```python
Table1
```

```python
Table2
```

```python
Table1.join(Table2)  # inner join
```

```python
Table1.join(Table2,how='outer')
```
