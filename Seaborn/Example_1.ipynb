```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event")
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri,hue="event",style="event",markers=True)
plt.show()
```

```python
import pandas as pd
```

```python
pokemon=pd.read_csv('pokemon.csv')
```

```python
pokemon.head()
```

```python
sns.barplot(x="is_legendary", y="speed", data=pokemon)
plt.show()
```

```python
sns.barplot(x="is_legendary",y="weight_kg",data=pokemon)
plt.show()
```

```python
sns.barplot(x="is_legendary",y="speed",hur="generation",data=pokemon)
plt.show()
```

```python
sns.barplot(x="is_legendary",y="weight_kg",data=pokemon,palette="vlag")  # palette="rocket" or palette="Blues_d"
plt.show()
```

```python
sns.barplot(x="is_legendary",y="weight_kg",data=pokemon,color="orange")
plt.show()
```
