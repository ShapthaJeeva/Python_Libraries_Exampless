```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = 2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.plot(x,y,color='r',linestyle=':',linewidth=4)
plt.title('Line Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x = np.arange(1,11)
y1 = 2*x
y2 = 3*x
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth=3)
plt.plot(x,y2,color='orange',linestyle='-',linewidth=5)

plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)  # 1 row, 2 columns, index 1
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)
plt.grid(True)

plt.show()
```

```python
plt.subplot(2,1,1)  # 1 row, 2 columns, index 1
plt.plot(x,y1,color='g',linestyle=':',linewidth=2)

plt.subplot(2,1,2)
plt.plot(x,y2,color='r',linestyle='-',linewidth=3)

plt.show()
```

```python
# Bar plot

student = {'Bob':87,'Matt':56,'Sam':27}
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
# Vertical bar plot
plt.bar(names,values)
plt.show()
```

```python
plt.bar(names,values)
plt.title('BMarks of Students')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
# Horizontal bar plot
plt.barh(names,values,color='g')
plt.title('Marks of Students')
plt.xlabel('Marks')
plt.ylabel('Students')
plt.grid(True)
plt.show()
```

```python
# Scatter plot

x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]  # both arrays should have same number of values

plt.scatter(x,a)
plt.show()
```

```python
plt.scatter(x,a,marker='*',c='g',s=100)  # s-size, c/color
plt.show()
```

```python
x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
b = [7,2,5,6,9,1,4,5,3]

plt.scatter(x,a,marker='*',c='r',s=100)
plt.scatter(x,b,marker='.',c='g',s=100)

plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,a,marker='*',c='g',s=100)

plt.subplot(1,2,2)
plt.scatter(x,b,marker='.',c='r',s=100)

plt.show()
```

```python
plt.subplot(2,1,1)
plt.scatter(x,a,marker='*',c='g',s=100)

plt.subplot(2,1,2)
plt.scatter(x,b,marker='.',c='r',s=100)

plt.show()
```

```python
# Box plot
# It gives 5 number summary(minimum value, 25% value, 50% value, 75% value, maximum value)

one = [1,2,3,4,5,6,7,8,9]
two = [1,2,3,4,5,4,3,2,1]
three = [6,7,8,9,8,7,6,5,4]
four = [7,8,9,6,8,6,5,8,9]

data = list([one,two,three,four])
```

```python
plt.boxplot(data)
plt.show()
```

```python
# Violin plot

plt.violinplot(data,showmedians=True)
plt.show()
```

```python
# Pie chart

fruit = ['Apple','Orange','Mango','Guava']
quantity = [67,34,100,29]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.2f%%',colors=['yellow','grey','blue','black'])
plt.show()
```

```python
# DoughNut chart
fruit = ['Apple','Orange','Mango','Guava']
quantity = [67,34,100,29]
```

```python
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors=['w'],radius=1)  # pie([1] <= we can give any random value here
plt.show()
```

```python
plt.pie(quantity,labels=fruit,radius=2,autopct='%0.2f%%',colors=['yellow','grey','blue','black'])
plt.pie([1],colors=['w'],radius=1)  # pie([1] <= we can give any random value here
plt.show()
```
